```python
fig = plt.figure(figsize = (10,6))
ax1 = fig.add_subplot(2,2,1)

ax2 = fig.add_subplot(2,2,2)

ax3 = fig.add_subplot(2,2,3)

#Create histogram
ax1.hist(wine.alcohol)

#create scatterplot
ax3.scatter(wine.alcohol,wine.deaths_heart)
```

```python
fig,ax = plt.subplots(2,2)
```

```python
fig,ax = plt.subplots(2,2)
ax[0][0].hist(wine.alcohol)
ax[1][1].scatter(wine.alcohol, wine.deaths_heart)
```

```python
fig,ax = plt.subplots(2,1)
ax[0].scatter(wine.alcohol, wine.deaths_heart)
ax[1].scatter(wine.alcohol, wine.deaths_liver)

ax[0].set_title("Heart Disease")
ax[1].set_title("Liver Disease")

fig.subplots_adjust(hspace = 0.7)
```

```python
ed = pd.read_csv("ed.csv")
```

```python
ed.head()
```

```python
new_ed = ed.set_index("Year")
```

```python
new_ed.head()
```

```python
new_ed.columns = ["no_hs","hs","some_coll","associates","bachelors","masters","professional","doctoral"]
```

```python
plt.plot(new_ed.masters)
```

```python
plt.plot(new_ed.masters,color="navy",linestyle="--")
plt.plot(new_ed.professional,color="k",linestyle=":")
plt.plot(new_ed.bachelors,color="blue",linestyle="dotted")

plt.legend(["masters","professional","bachelors"])
```

```python
plt.plot(new_ed.masters,color="navy")
plt.plot(new_ed.professional,color="k")
plt.plot(new_ed.bachelors,color="blue")
plt.plot(new_ed.doctoral,color="red")
plt.plot(new_ed.associates,color="pink")
plt.plot(new_ed.some_coll,color="green")
plt.plot(new_ed.hs,color="orange")
plt.plot(new_ed.no_hs,color="brown")

plt.legend(["masters","professional","bachelors","doctoral","associates","some_coll","hs","no_hs"])
plt.legend(bbox_to_anchor=(1,.95))
plt.xlabel("Years")
plt.ylabel("Unemployment Rate" )
plt.title("Degrees & Unemployment")
```
