```python
from nltk import FreqDist
from nltk.tokenize import word_tokenize
from nltk.corpus import inaugural
```

```python
print(inaugural.fileids())
#inaugural speeches of all presidents
```

```python
wash1 = inaugural.raw("1789-Washington.txt")
print(wash1)
```

```python
len(wash1)
```

```python
#word frequency distribution
fdist = FreqDist(wash1)
print(fdist.most_common(20))
```

```python
print(fdist.plot(20))
```

```python
wash1_words = inaugural.words("1789-Washington.txt")
print(wash1_words)
```

```python
fdist = FreqDist(wash1_words)
print(fdist.plot(20))
```

```python
#stop words are those commonly used words that you don't care about
from nltk.corpus import stopwords
stopwords = nltk.corpus.stopwords.words("English")
print(stopwords)
```

```python
type(stopwords)
```

```python
clean_wash1 = []
for item in wash1_words:
    if item not in stopwords:
        clean_wash1.append(item)
print(clean_wash1)
```

```python
fdist2 = FreqDist(clean_wash1)
x = fdist2.most_common(20)
print(x)
print(fdist2.plot(20))
```

```python
new_stop_words = [",",".","I",";","-","In","may"]
stopwords.extend(new_stop_words)
clean_wash2 = []
for item in wash1_words:
    if item not in stopwords:
        clean_wash2.append(item)
print(clean_wash2)
```

```python
fdist3 = FreqDist(clean_wash2)
x = fdist3.most_common(20)
print(x)
print(fdist3.plot(20))
```

```python
#counting bigrams as opposed to unigrams
fdistbi1 = list(nltk.bigrams(clean_wash2))
fdist4 = FreqDist(fdistbi1)
print(fdist4.most_common(20))
print(fdist4.plot(20))
```

```python
from nltk.corpus import inaugural


cfd = nltk.ConditionalFreqDist((target,fileid[:4])
                               for fileid in inaugural.fileids()
                               for word in inaugural.words(fileid)
                               for target in["liberty","equality","brotherhood"]
                               if word.lower().startswith(target))
cfd.plot(title="French ideals in US American presidential speeches")
```
