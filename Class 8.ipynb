```python
#filtering on pandas data
#Query is a basic one 
df.query("speeding>7")
```

```python
df.query("speeding==9.45")
```

```python
df.query("speeding>7 and speeding>9")
```

```python
#You can use both and and & to make it work in the query method
df.query("speeding>7 & speeding>9")
```

```python
df.query("speeding>7 or speeding>9")
```

```python
#You have to put a string withhin its own set of parenthesis to pull a string
df.query("abbrev=='PA'")
```

```python
#Don't use a list here, it's a panda data object
df.query("abbrev not in ('AL','AK','AR','AZ','WA','WY','WV','WI')")
```

```python
#I missed commas between AK, AR, and AZ here, and it doesn't throw an error, but just prints it
df.query("abbrev not in ('AL','AK''AR''AZ','WA','WY','WV','WI')")
```

```python
#To pull a variable within the query you use the @ symbol to call it
states = ['AL','AK','AR','AZ','WA','WY','WV','WI']
df.query("abbrev not in @states")
```
